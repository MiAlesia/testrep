## Visualizing distribution of data

Plot a histogram of `ConvertedComp.`

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

Query3="""
select Age,WorkWeekHrs from master"""

data_3=pd.read_sql_query(Query3,conn)

data_3.plot(kind='scatter', x='Age',y='WorkWeekHrs',figsize=(10,8))

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

Query4 = """
select WorkWeekHrs,CodeRevHrs,Age from master"""
data_4=pd.read_sql_query(Query4,conn)
data_4.head()
data_4.shape
```

```python
data_4_new = data_4.dropna()
data_4_new.shape
```

```python
data_4_new.head()
```

```python
norm_W = (data_4_new['WorkWeekHrs'] - data_4_new['WorkWeekHrs'].min())/(data_4_new['WorkWeekHrs'].max()-data_4_new['WorkWeekHrs'].min())
norm_C = (data_4_new['CodeRevHrs'] - data_4_new['CodeRevHrs'].min())/(data_4_new['CodeRevHrs'].max()-data_4_new['CodeRevHrs'].min())

ax0= data_4_new.plot(kind='scatter',x='Age',y='WorkWeekHrs',figsize=(15,10),alpha=0.4,color='orange',xlim=(18,100),s=norm_W*200)
ax1= data_4_new.plot(kind='scatter',x='Age',y='CodeRevHrs',figsize=(15,10),alpha=0.4,color='blue',s=norm_C*200,ax=ax0)

ax0.legend(['WorkWeekHrs', 'CodeRevHrs'],loc = 'upper right', fontsize='x-large')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
Query5="""
select count(*) as Num_Of_Resp,DatabaseDesireNextYear from DatabaseDesireNextYear group by DatabaseDesireNextYear order by count(*) desc limit 5"""
data_5 = pd.read_sql_query(Query5,conn)

data_5
```

```python
data_5['Num_Of_Resp'].plot(kind='pie',
                           figsize=(8,8),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=False,
                           labels=data_5['DatabaseDesireNextYear'],
                           colors=['gold','lightblue','orange','lightgreen','pink'])

plt.title('Top 5 Databases Respondents wish to learn next year')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query6 = """
select WorkWeekHrs,CodeRevHrs, Age
from master
where Age >=30 and Age <=35
"""
data_6_1=pd.read_sql_query(Query6,conn)
data_6 = data_6_1.groupby('Age').median()
data_6
```

```python
labels = ['30', '31', '32', '33', '34','35']
width = 0.4      # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(10,5))

ax.bar(labels,data_6['WorkWeekHrs'], width, label='WorkWeekHrs')
ax.bar(labels, data_6['CodeRevHrs'], width, label='CodeRevHrs')

ax.set_ylabel('WorkWeekHrs and CodeRevHrs')
ax.set_xlabel('Age')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
ax.legend()
```

```python
data_6.plot(kind='bar',
            figsize=(10,5),
            stacked=True)

plt.title('Median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.ylabel('Hours')
plt.legend(loc = 'upper right', fontsize='x-large')
```
